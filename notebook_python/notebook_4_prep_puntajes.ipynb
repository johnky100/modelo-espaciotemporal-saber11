```python
import os
import pandas as pd
```

```python
# RUTAS
```

```python

# RUTAS

ruta_base = r"C:/Users/john/Desktop/Saber_11_2025/data/4_modelado"
ruta_salida = r"C:/Users/john/Desktop/Saber_11_2025/resultados/puntajes"

os.makedirs(ruta_salida, exist_ok=True)

print("Rutas listas")
print(" - Entrada:", ruta_base)
print(" - Salida :", ruta_salida)
```

```python
# ARCHIVOS EXACTOS A PROCESAR
```

```python
# ARCHIVOS EXACTOS A PROCESAR

archivos_objetivo = {
    2015: "imp2015_db1_completo.csv",
    2016: "imp2016_db1_completo.csv",
    2017: "imp2017_db1_completo.csv",
    2018: "imp2018_db1_completo.csv",
    2019: "imp2019_db1_completo.csv",
    2020: "imp2020_db1_completo.csv",
    2021: "imp2021_db1_completo.csv",
    2022: "imp2022_db1_completo.csv"
}
```

```python
# LISTA PARA GUARDAR LOS RESULTADOS
```

```python

# LISTA PARA GUARDAR LOS RESULTADOS

lista_resultados = []
```

```python
# PROCESAMIENTO POR AÑO
```

```python
# PROCESAMIENTO POR AÑO

for year, archivo in archivos_objetivo.items():

    print(f"=== PROCESANDO {year} ===")

    path_file = os.path.join(ruta_base, archivo)

    if not os.path.exists(path_file):
        print(f"NO existe el archivo esperado: {archivo}")
        continue

    print(f"Archivo usado: {archivo}")

    # Leer archivo
    df = pd.read_csv(path_file, low_memory=False)

    print("Shape:", df.shape)

    # Columna departamental

    col_dep = "COLE_DEPTO_UBICACION"

    if col_dep not in df.columns:
        raise KeyError(f"No existe la columna {col_dep} en {archivo}")

    print(f"Columna {col_dep} encontrada.")
    print("Primeros valores:", df[col_dep].head().tolist())
    print("Valores únicos (10):", df[col_dep].unique()[:10])

    # AGRUPACIÓN POR DEPARTAMENTO
    
    df_dep = (
        df.groupby(col_dep, as_index=False)
          .agg(PUNTAJE_GLOBAL_PROMEDIO=("PUNT_GLOBAL", "mean"))
    )

    df_dep["ANIO"] = year

    print("== RESULTADO DEL AÑO ==")
    print(df_dep.head())
    print(df_dep.shape)

    lista_resultados.append(df_dep)
```

```python
# UNIR TODOS LOS AÑOS
```

```python

# UNIR TODOS LOS AÑOS

df_final = pd.concat(lista_resultados, ignore_index=True)

print("DATAFRAME FINAL — PUNTAJES 2015–2022")

print(df_final.head())
print(df_final.tail())
print("Shape final:", df_final.shape)
```

```python
# EXPORTACIÓN FINAL
```

```python
# EXPORTACIÓN FINAL

output_csv = os.path.join(ruta_salida, "puntajes_agregados_2015_2022.csv")
output_parquet = os.path.join(ruta_salida, "puntajes_agregados_2015_2022.parquet")

df_final.to_csv(output_csv, index=False)
df_final.to_parquet(output_parquet)

print("\nArchivos exportados:")
print(" - CSV    :", output_csv)
print(" - Parquet:", output_parquet)
```
